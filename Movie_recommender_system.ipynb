```python
#Fetching the useful features from movies dataframe
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#checking null values in dataframe
movies.isnull().sum()
```

```python
#droping the nll values row
movies.dropna(inplace=True)
```

```python
#checking duplicate row 
movies.duplicated().sum()
```

```python
#Fetching genre name from genres feature and storing in list
import ast
def convert_to_list(obj):
    genre_list=[]
    for i in ast.literal_eval(obj):
        genre_list.append(i['name'])
    return genre_list
```

```python
movies['genres']=movies['genres'].apply(convert_to_list)
```

```python
movies
```

```python
#Fetching name keyword from keyword feature and storing in list
movies['keywords']=movies['keywords'].apply(convert_to_list)
```

```python
movies
```

```python
#Fetching three top actor name from cast feature
import ast
def actor(obj):
    
    actor_name=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            actor_name.append(i['name'])
            count+=1
        else:
            break
    return actor_name

        
```

```python
movies['cast']=movies['cast'].apply(actor)
```

```python
#Fetching director name from crew feature
import ast
def fetch_director(obj):
    director_name=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            director_name.append(i['name'])
            break
    return director_name
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#converting into list of overview feature
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
#joining two word which is seperated by space
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
#concatinating overview,genres,keywords,cast,crew features into 'tags' feature
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#creating new dataframe with movie_id and tags feature
df=movies[['movie_id','title','tags']]
```

```python
df.head()
```

```python
#joining tag feature with space
df['tags']=df['tags'].apply(lambda x:' '.join(x))
```

```python
df.head(1).values
```

```python
#converting lower case of tag feature
df['tags']=df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
#stemming the repetitive word(like actors,actor converted to actor word)
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    list_stem=[]
    for i in text.split():
        list_stem.append(ps.stem(i))
    return ' '.join(list_stem)
        
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
#converting tag feature text into vector using bag of words with CountVectorizer scikit learn library
#max_features is 5000 words which frequently occur in all tags feature
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
#words which occur most frequently in tags feature
cv.get_feature_names()
```

```python
#calculating cosine similarity between the vector
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=df[df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```
